```python
from sklearn.preprocessing import StandardScaler
```

```python

X = df.drop(['Weekly_Sales', 'Date'], axis=1)
y = df['Weekly_Sales']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, 
                                                   random_state=101)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()  # Adjust hyperparameters as needed
regressor.fit(X_train, y_train)

```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Mean Squared Error:", rmse)
```

```python
def visualize_tree(regressor, feature_names):
    plt.figure(figsize=(20, 10))  # Set the figure size
    plot_tree(regressor, filled=True, feature_names=feature_names, class_names=True)  # Plot the decision tree
    plt.show()
```

```python
visualize_tree(regressor,"Weekly_Sales")  # Assuming last column is the target variable
```
